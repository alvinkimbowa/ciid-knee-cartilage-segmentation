```python
img = "../../Datasets/raw_datasets/5.1 clarius_auto_msu_parmar_segmentations/images/Nike_s500_v6_CAR_R1.png"
mask = "../../Datasets/raw_datasets/5.1 clarius_auto_msu_parmar_segmentations/masks/Nike_s500_v6_CAR_R1_mask.png"

img = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
mask = cv2.imread(mask, cv2.IMREAD_GRAYSCALE)

print("img: ", img.max())
print("mask: ", mask.shape)

plt.figure(figsize=(20,5))
plt.imshow(img, cmap='gray')
plt.show()

plt.figure(figsize=(20,5))
plt.imshow(np.multiply(img, np.int32(img>250)), cmap='gray')
plt.show()
```
